```python
from collections import Counter
import numpy as np

corpus = [
    "I love machine learning",
    "Machine learning is powerful",
    "I love deep learning"
]

# Preprocess: lowercase + split
processed_docs = [doc.lower().split() for doc in corpus]

# Create vocabulary
vocab = sorted(set(word for doc in processed_docs for word in doc))
print("Vocabulary:", vocab)

# Create BoW matrix
bow_matrix = []

for doc in processed_docs:
    word_count = Counter(doc)
    bow_matrix.append([word_count[word] for word in vocab])

bow_matrix = np.array(bow_matrix)

print("\nBag of Words Matrix:\n")
print(bow_matrix)
```

```python
import seaborn as sns
```

```python
sns.heatmap(bow_matrix, annot=True, cmap='Blues',xticklabels=vocab,yticklabels=['Doc1', 'Doc2', 'Doc3'])
```

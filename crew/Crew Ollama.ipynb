```python
from crewai import Agent, Task, Crew, Process
from crewai.llm import LLM
```

```python
ollama_llm = LLM(
    model="ollama/llama3",
    base_url="http://localhost:11434"
)
```

```python
researcher = Agent(
    role="Researcher",
    goal="Research the topic thoroughly",
    backstory="You gather accurate technical information.",
    llm=ollama_llm,
    verbose=True
)
```

```python
writer = Agent(
    role="Technical Writer",
    goal="Convert research into simple language",
    backstory="You explain technical concepts clearly.",
    llm=ollama_llm,
    verbose=True
)
```

```python
research_task = Task(
    description="Research how CrewAI integrates with Ollama.",
    expected_output="Key technical points and workflow.",
    agent=researcher
)
```

```python
writing_task = Task(
    description="Write a beginner-friendly explanation based on the research.",
    expected_output="A simple and clear article.",
    agent=writer
)
```

```python
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writing_task],
    process=Process.sequential
)
```

```python
result = crew.kickoff()
print(result)
```

```python
from crewai.llm import LLM

llm = LLM(
    model="gemini/gemini-1.5-flash",
    api_key=os.environ["GEMINI_API_KEY"],
    temperature=0.2
)
```

```python
from crewai import Agent
```

```python
research_agent = Agent(
    role="AI Research Analyst",
    goal="Research and explain technical concepts accurately",
    backstory="Specialist in LLM frameworks and multi-agent systems",
    llm=llm,
    verbose=True
)
```

```python
writer_agent = Agent(
    role="Technical Writer",
    goal="Convert research into clear and structured explanations",
    backstory="Expert at simplifying complex AI topics for developers",
    llm=llm,
    verbose=True
)
```

```python
from crewai import Task
```

```python
research_task = Task(
    description="Research CrewAI and how multi-agent systems work",
    expected_output="Detailed technical notes on CrewAI",
    agent=research_agent
)
```

```python
writing_task = Task(
    description=(
        "Using the research notes, write a clear explanation of CrewAI "
        "and multi-agent systems for beginners"
    ),
    expected_output="Well-structured beginner-friendly explanation",
    agent=writer_agent
)
```

```python
from crewai import Crew
```

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, writing_task]
)

result = crew.kickoff()
print(result)
```

```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
data = {
    "text":[
        "this movie is very good",
        "i love this film",
        "this movie is amazing",
        "i hate this movie",
        "this film is bad",
        "very boring movie"
    ],
    
    "label":[1,1,1,0,0,0]   # 1 = Positive , 0 = Negative
}

df = pd.DataFrame(data)
```

```python
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df['text'])

sequences = tokenizer.texts_to_sequences(df['text'])
```

```python
X = pad_sequences(sequences, maxlen=6)
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = Sequential()

model.add(Embedding(input_dim=5000, output_dim=64, input_length=6))

model.add(LSTM(64))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=2)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy:", accuracy)
```

```python
text = ["this movie is good"]

seq = tokenizer.texts_to_sequences(text)

padded = pad_sequences(seq, maxlen=6)

prediction = model.predict(padded)

if prediction > 0.5:
    print("Positive Sentiment")
else:
    print("Negative Sentiment")
```

```python
text = ["very boring movie"]

seq = tokenizer.texts_to_sequences(text)

padded = pad_sequences(seq, maxlen=6)

prediction = model.predict(padded)

if prediction > 0.5:
    print("Positive Sentiment")
else:
    print("Negative Sentiment")
```

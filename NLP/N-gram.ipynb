```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
text = [
    """Deep learning is a subset of machine learning that uses artificial neural networks with multiple 
    hidden layers to model complex patterns in data. It is inspired by the structure and function of the 
    human brain, particularly biological neurons and synaptic connections.These deep neural networks 
    automatically learn hierarchical feature representations from raw input data, eliminating the need 
    for manual feature engineering. This makes deep learning especially powerful for high-dimensional 
    data such as images, audio, and text.""",

    """Neural network consists of layers of interconnected nodes or neurons that collaborate to process 
    input data. In a fully connected deep neural network data flows through multiple layers where each 
    neuron performs nonlinear transformations, allowing the model to learn intricate representations of the data.
    In a deep neural network the input layer receives data which passes through hidden layers that transform the 
    data using nonlinear functions. The final output layer generates the model’s prediction.""",

    """Deep learning architectures such as Convolutional Neural Networks (CNNs) and Recurrent Neural Networks 
    (RNNs) are widely used in computer vision and natural language processing. More advanced models like 
    Transformers have further improved performance in sequence modeling tasks.Deep learning requires large 
    datasets, significant computational power, and optimization algorithms like gradient descent with 
    backpropagation. It is widely applied in areas such as speech recognition, autonomous vehicles, medical 
    diagnosis, and recommendation systems."""
]
text
```

```python
text_words = sum(len(p.split()) for p in text)
print("Total number of words in text:", text_words)
```

```python
#UNIGRAM MODEL
print("UNIGRAM MODEL")

unigram_vectorizer = CountVectorizer(ngram_range=(1, 1),stop_words='english')
X_uni = unigram_vectorizer.fit_transform(text)

print("\nUnigram Vocabulary:")
print(unigram_vectorizer.get_feature_names_out())

print("\nUnigram Matrix:")
print(X_uni.toarray())

uni_similarity = cosine_similarity(X_uni)

print("\nUnigram Cosine Similarity Matrix:")
print(uni_similarity)
```

```python
#BIGRAM MODEL

bigram_vectorizer = CountVectorizer(ngram_range=(2, 2),stop_words='english')
X_bi = bigram_vectorizer.fit_transform(text)

print("\nBigram Vocabulary:")
print(bigram_vectorizer.get_feature_names_out())

print("\nBigram Matrix:")
print(X_bi.toarray())

bi_similarity = cosine_similarity(X_bi)

print("\nBigram Cosine Similarity Matrix:")
print(bi_similarity)
```

```python
#TRIGRAM MODEL

trigram_vectorizer = CountVectorizer(ngram_range=(3, 3),stop_words='english')
X_tri = trigram_vectorizer.fit_transform(text)

print("\nTrigram Vocabulary:")
print(trigram_vectorizer.get_feature_names_out())

print("\nTrigram Matrix:")
print(X_tri.toarray())

tri_similarity = cosine_similarity(X_tri)

print("\nTrigram Cosine Similarity Matrix:")
print(tri_similarity)
```

```python
#UNIGRAM + BIGRAM MODEL

uni_bi_vectorizer = CountVectorizer(ngram_range=(1, 2),stop_words='english')
X_uni_bi = uni_bi_vectorizer.fit_transform(text)

print("\nUnigram + Bigram Vocabulary:")
print(uni_bi_vectorizer.get_feature_names_out())

print("\nUnigram + Bigram Matrix:")
print(X_uni_bi.toarray())

uni_bi_similarity = cosine_similarity(X_uni_bi)

print("\nUnigram + Bigram Cosine Similarity Matrix:")
print(uni_bi_similarity)
```

```python
#BIGRAM + TRIGRAM MODEL

bi_tri_vectorizer = CountVectorizer(ngram_range=(2, 3),stop_words='english')
X_bi_tri = bi_tri_vectorizer.fit_transform(text)

print("\nBigram + Trigram Vocabulary:")
print(bi_tri_vectorizer.get_feature_names_out())

print("\nBigram + Trigram Matrix:")
print(X_bi_tri.toarray())

bi_tri_similarity = cosine_similarity(X_bi_tri)

print("\nBigram + Trigram Cosine Similarity Matrix:")
print(bi_tri_similarity)
```

```python
print("Unigram Cosine Similarity Matrix:")
print(uni_similarity)

print("Bigram Cosine Similarity Matrix:")
print(bi_similarity)

print("Trigram Cosine Similarity Matrix:")
print(tri_similarity)

print("Unigram + Bigram Cosine Similarity Matrix:")
print(uni_bi_similarity)

print("Bigram + Trigram Cosine Similarity Matrix:")
print(bi_tri_similarity)
```
